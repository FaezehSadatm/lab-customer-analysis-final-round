```python
def log_transform(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

### Processing Data

```python
def boxcox_transform(df):
    numeric_cols = df.select_dtypes(np.number).columns
    _ci = {column: None for column in numeric_cols}
    for column in numeric_cols:
        df[column] = np.where(df[column]<=0, np.NAN, df[column]) 
        df[column] = df[column].fillna(df[column].mean())
        transformed_data, ci = stats.boxcox(df[column])
        df[column] = transformed_data
        _ci[column] = [ci] 
    return df, _ci
```

```python
data, _ci = boxcox_transform(df)
data
```

```python
for column in data.select_dtypes(np.number).columns:
    sns.distplot(data[column])
    plt.show()
```

```python
def remove_outliers(df, threshold=1.5, in_columns=data.select_dtypes(np.number).columns, skip_columns=[]):
    for column in in_columns:
        if column not in skip_columns:
            upper = np.percentile(df[column],75)
            lower = np.percentile(df[column],25)
            iqr = upper - lower
            upper_limit = upper + threshold * iqr
            lower_limit = lower - threshold * iqr
            df = df[(df[column]>lower_limit) & (df[column]<upper_limit)]
    return df
```

```python
data = remove_outliers(data, threshold=1.5, in_columns=['INCOME'])
```

### Train and test split

```python
y = data['TOTAL_CLAIM_AMOUNT']
X = data.drop(['TOTAL_CLAIM_AMOUNT'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = 'object')
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)
```

```python
encoded = encoder.transform(X_cat).toarray()
X = np.concatenate([X_num, encoded], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Apply the Model
